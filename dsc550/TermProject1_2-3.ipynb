##### Removing the outliers
Removing few outliers as the price of houses might not reflect the actual price change. for example,outliers such as newer houses or oldest ones

```python
q_low = df.quantile(0.10)
q_hi  = df.quantile(0.90)
df_filtered = df[(df < q_hi) & (df > q_low)]
```

##### Implementing Linear regression 

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop(['SALE_PRC'],axis=1)
y = df['SALE_PRC']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# printing the shapes of the new X objects
print(X_train.shape)
print(X_test.shape)
 
# printing the shapes of the new y objects
print(y_train.shape)
print(y_test.shape)
```

```python
# training the model on training set
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
 
# making predictions on the testing set
y_pred = knn.predict(X_test)
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("kNN model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.pipeline import Pipeline,FeatureUnion
```

```python
# define min max scaler
scaler = MinMaxScaler()
```

```python
knn = KNeighborsClassifier(n_neighbors=5,n_jobs=-1)
```

```python
pipeline = Pipeline([("standardizer",scaler),("classifier",KNeighborsClassifier())])
```

```python
pipeline.fit(X_train,y_train)
```

```python
pipeline.score(X_test,y_test)
```
